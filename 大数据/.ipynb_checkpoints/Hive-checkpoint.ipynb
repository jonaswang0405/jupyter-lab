### Hive的MetaStore多次启动失败，提示端口被占用解决办法：
##### 执行命令：netstat -apn|grep -i "listen"，开启监听所有端口，找到被占用的端口
##### 然后执行：kill -9 进程ID，然后再执行hive --service metastore即可成功
##### 方案2：执行给该命令，直接以一个新的端口（10001）号启动 hive --service metastore -p 10001 &

### Hive窗口函数：
##### concat_list(colName):将一列中的多行，合并为一行。含重复值
##### concat_set(colName)：将一列中的多行，合并为一行。不含重复值
##### concat_ws("-","it","abc"):拼接多个字符串，并可指定分隔符
##### partition：rows between 从前3行至当前行
##### select  create_time,sum(cash) over(partition by province order by create_time rows between 3 preceding and current row) as sum_case from table
##### partition：rows between 从前3行至下一行
##### select  create_time,sum(cash) over(partition by province order by create_time rows between 3 preceding and 1 following) as sum_case from table
### 排序：
##### row_number：每个分组中，每行分配一个从1开始的唯一序号，递增，不考虑重复；1，2，3，3，5
##### rank：每个分组中，每行分配一个从1开始的唯一序号，考虑重复，挤占后续排位；1，2，3，3，4
##### dense_rank: 每个分组中，为每行分配一个从1开始的序号，考虑重复，不基站后续位置；1，2，3，4，5
##### row_number() over(partition by province order by create_time desc)
##### ntile：将分组内的数据指定为若干个组
##### ntile(3) over(partition by province order by create_time)

##### Lag(col,n,default):用以统计窗口内前n行的值，default，如遇到空值，则为该默认值
##### Lead(col,n,default):用以统计窗口内往下第n行的值
##### First_Value:取分组排序后，截至当前行第一个值
##### Last_Value:取分组内排序后，截至当前行最后一个值
##### lag(cash,1) over (partition by create_time)

### Lateral view
