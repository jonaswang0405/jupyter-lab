hadoop虚拟机--分布式部署--免密登录：

```python
免密登录：
1、在各主机上面执行：ssh-keygen -t rsa，一路回车。
2、cd .ssh/ 切换到ssh目录
3、cat id_rsa.pub >> authorized_keys 将公钥复制到authorized_keys文件夹（本身不存在，自动创建）。
4、对authorized_keys文件进行授权：
   chmod 700 ~/.ssh
   chmod 644 ~/.ssh/authorized_keys
5、将authorized_keys文件复制到各服务器，每个里面均执行cat id_rsa.pub >> authorized_keys.
  确保每个服务器的公钥都在每个authorized_keys文件内，id_rsa.pub即公钥文件，可查看里面内容。
6、然后将把包含所有服务器公钥的authorized_keys复制到各服务器一份：
  scp ~/.ssh/authorized_keys slave1:~/.ssh/  。
7、再分别将每个服务器上的authorized_keys复制到其他服务器上的.ssh/目录下
```
